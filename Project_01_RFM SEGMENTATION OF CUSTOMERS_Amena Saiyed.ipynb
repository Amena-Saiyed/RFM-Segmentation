# Task 1 : Create 3 different dataframes(recent, frequent, monetary) from your complete customer dataframe.

```python
#Getting the data
import pandas as pd
df = pd.DataFrame()
df = pd.read_csv('customer_invoices_data.csv')
df
```

```python
#Cleaning the data
df = df.dropna()
df
```

```python
#import modules
import pandas as pd # for dataframes
import matplotlib.pyplot as plt # for plotting graphs
import seaborn as sns # for plotting graphs
import datetime as dt

#taking current date
PRESENT = dt.datetime(2022,12,8)

#formatting InvoiceDate in the dataset 
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df
```

```python
#Calculating R,F & M for each Customer ID
rfm= df.groupby('Customer ID').agg({'InvoiceDate': lambda date: (PRESENT - date.max()).days,
                                        'Invoice': lambda num: len(num),
                                        'Price': lambda price: price.sum()})
# Change the name of columns
rfm.rename(columns = {'InvoiceDate':'Recency(days)', 'Invoice':'Frequency',
                              'Price':'Monetary'}, inplace = True)
rfm
```

# Task 2 : Divide the customers group data into 4 items for R F & M respectively.

```python
#Dividing dataframes into bins(categorizing data)
rfm['r_quartile'] = pd.qcut(rfm['Recency(days)'], 4, ['Most Recent','Above Average','Below Average','Least Recent'] )
rfm['f_quartile'] = pd.qcut(rfm['Frequency'], 4, ['Most Frequent','Above Average','Below Average','Least Frequent'])
rfm['m_quartile'] = pd.qcut(rfm['Monetary'], 4, ['Most Monetary','Above Average','Below Average','Least Monetary'])
```

```python
#Making a new column 'RFM_Score' where the values for R,F & M after categorizing will be merged as string and can be used
#further for customer segmentation.
rfm['RFM_Score'] = rfm.r_quartile.astype(str) + rfm.f_quartile.astype(str) + rfm.m_quartile.astype(str)
rfm
```

# Task 3 : Assign segment based upon your division in Task 2

```python
import  numpy as np
rfm["Customer_segment"] = np.where(rfm['RFM_Score'] =='Most RecentMost FrequentMost Monetary', "Best Customers",
                          (np.where(rfm['RFM_Score'] == 'MostRecentMostFrequentMostMonetary',"High Spending Customer",
                          (np.where(rfm['RFM_Score'] == 'Below AverageBelow AverageBelow Average',"Low Spending Customer",
                           np.where(rfm['RFM_Score'] == 'Least RecentLeast FrequentLeast Monetary','Churned Best Customers', 
                                    'Inactive Customers'))))))
rfm
```

```python
rfm['r_quartile'] = pd.qcut(rfm['Recency(days)'], 4, ['1','2','3','4'] )
rfm['f_quartile'] = pd.qcut(rfm['Frequency'], 4, ['1','2','3','4'] )
rfm['m_quartile'] = pd.qcut(rfm['Monetary'], 4, ['1','2','3','4'] )
```

```python
rfm['RFM_Score'] = rfm.r_quartile.astype(str) + rfm.f_quartile.astype(str) + rfm.m_quartile.astype(str)
rfm
```

```python
import  numpy as np
rfm["Customer_segment"] = np.where(rfm['RFM_Score'] =='111', "Best Customers",
                          (np.where(rfm['RFM_Score'] == '222',"High Spending Customer",
                          (np.where(rfm['RFM_Score'] == '333',"Low Spending Customer",
                           np.where(rfm['RFM_Score'] == '444','Churned Best Customers', 
                                    'Inactive Customers'))))))
rfm
```
